# Sea Extent and Temperature Anomalies
Sea extent in million square kilometers is the area of ocean where at least 15% of the surface is frozen.

Temperatude anomalies in degrees Celcius is the deviation of temperature from the average temperature. Positive anomalies show that the observed temperature was warmer than the baseline, whereas a negative anomaly indicates that the observed temperature was lower than baseline.

```python
# plot the data
# create line of best fit
# write the equation of the line
# do a prediction for 2030
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

sns.set()
plt.figure(figsize=(15, 7))
plt.scatter(data['Temperature Anomaly'], data['Sea Ice Extent'], color = 'black', alpha=0.7)

for i in range(len(data)):
    plt.text(data['Temperature Anomaly'][i], data['Sea Ice Extent'][i], data['Year'][i], fontsize=13, ha='center', va='top')
    
plt.xlabel('Temperature Anomaly (in degrees Celcius)', fontsize = 15, color='black')
plt.ylabel('Sea Ice Extent (in million square killometers)', fontsize = 15, color='black')
plt.xticks(fontsize=13, color='black')
plt.yticks(fontsize=13, color='black')
plt.title('Temperature Anomaly vs Sea Ice Extent', fontsize = 15, color='black')

x = data['Temperature Anomaly'].values.reshape(-1, 1)
y = data['Sea Ice Extent']

poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
coef_main = model.coef_
intercept_main = model.intercept_

plt.plot(x, model.predict(x_poly), color='blue', linewidth=2)

# PolynomialFeatures to generate polynomial features up to degree 2, effectively transforming your original 
# feature space into a higher-dimensional space. Then, you're using LinearRegression from scikit-learn to fit 
# this transformed feature space. Despite fitting a quadratic curve, you're still using linear regression to 
# estimate the coefficients of this curve.
```

```python
def predict_val(temp):
    return coef_main[2] * (temp ** 2) + coef_main[1]*temp + intercept_main
```

```python
years = np.array(list(temp_anomalies_data['Temperature'].keys())).reshape(-1, 1)

# Assuming temp_anomalies_data['Temperature'].values contains temperatures
temps = np.array(list(temp_anomalies_data['Temperature'].values)).reshape(-1, 1)

# Fit linear regression model
model = LinearRegression()
model.fit(years, temps)

next_year = np.array([[2030]])
predicted_temp = model.predict(next_year)
temp_anom_2030 = predicted_temp[0][0]
```

```python
predict_val(temp_anom_2030) # predicted value for 2030
```

Flood data globally with economic loss in billion US dollars.

```python
flood_data
```

Determine what percentage of the total yearly economic loss was due to floods.

```python
# available flood data in China between 2013 to 2021
# 2016
flood_2016 = flood_data['Economic Loss'][8]

# 2020
flood_2020 = flood_data['Economic Loss'][5]

# 2021
flood_2021 = flood_data['Economic Loss'][9]
```

```python
# Total economic loss in China during the years 2016, 2020, and 2021
# 2016
total_2016 = economic_loss_data['Economic Loss (billion US dollars)'][2016]

# 2020
total_2020 =economic_loss_data['Economic Loss (billion US dollars)'][2020]

# 2023
total_2021 = economic_loss_data['Economic Loss (billion US dollars)'][2021]
```

```python
plt.figure(figsize=(15, 7))
data_2016 = [flood_2016, total_2016]
data_2020 = [flood_2020, total_2020]
data_2021 = [flood_2021, total_2021]

plt.bar(2016, data_2016[1], color = '#1f77b4')
plt.bar(2016, data_2016[0], color = 'orange')

plt.bar(2020, data_2020[1], color = '#1f77b4')
plt.bar(2020, data_2020[0], color = 'orange')

plt.bar(2021, data_2021[1], color = '#1f77b4')
plt.bar(2021, data_2021[0], color = 'orange')

plt.title('Economic Loss due to Flooding VS Total Economic Loss in China', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Economic Loss (in billions US dollars)', fontsize = 15)
plt.xticks(fontsize=13, color='black')
plt.yticks(fontsize=13, color='black')

plt.show()
```

```python
percent_2016 = round(data_2016[0]/data_2016[1]*100, 2)
percent_2020 = round(data_2020[0]/data_2020[1]*100, 2)
percent_2021 = round(data_2021[0]/data_2021[1]*100,2)

percent_arr = [percent_2016, percent_2020, percent_2021]
```

```python
# average of economic loss due to flooding
sum_per = 0
for percent in percent_arr:
    sum_per += percent
average_percent = round(sum_per / len(percent_arr), 2)
```

```python
average_percent
```

#### Answer:

On average, about 70% of yearly economic loss in China is the result of natural disasters.
