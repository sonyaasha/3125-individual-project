# Sea Extent and Temperature Anomalies
Sea extent in million square kilometers is the area of ocean where at least 15% of the surface is frozen.

Temperatude anomalies in degrees Celcius is the deviation of temperature from the average temperature. Positive anomalies show that the observed temperature was warmer than the baseline, whereas a negative anomaly indicates that the observed temperature was lower than baseline.

### 1. How have the anomalies in temperature impacted the sea ice extent?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Visualizing Sea Ice Extent Data

```python
sns.set()
plt.figure(figsize=(15, 7))

for index, row in sea_ice_extent_data.iterrows():
    plt.plot(row.index, row.values, label=index, linewidth = 2) 
    
plt.xticks(fontsize=13, color='black')
plt.yticks(fontsize=13, color='black')

plt.xlabel('Month', fontsize = 15, color='black')
plt.ylabel('Sea Ice Extent (million square kilometers)', fontsize = 15, color='black')
plt.title('Sea Ice Extent over the Year', fontsize = 15, color='black')
plt.legend(loc = 'best', fontsize = 15)
```

#### Visualizing Temperature Anomalies

```python
sns.set()
plt.figure(figsize=(15, 7))

plt.plot(temp_anomalies_data.index, temp_anomalies_data['Temperature'].values, color="#f39800", linewidth = 2)
    
plt.xticks(fontsize=13, color='black')
plt.yticks(fontsize=13, color='black')

plt.xlabel('Year', fontsize = 15, color='black')
plt.ylabel('Temperature Deviation (in degrees Celcius)', fontsize = 15, color='black')
plt.title('Temperature Deviation from Average', fontsize = 15, color='black')
```

### Analyzing years 1980, 1990, 2000, 2010, and 2023 for patterns

```python
model_year = []
model_temp_anomaly = []
model_average_sea_extent = []
```

#### Looking at 1980

```python
year = 1980
data_1980 = sea_ice_extent_data.loc[year] # to access data at specific year
temp_anom_1980 = temp_anomalies_data.loc[year]['Temperature']
```

```python
sns.set()
plt.figure(figsize=(15, 7))

plt.plot(data_1980.index, data_1980.values, label=index, linewidth = 2) 
    
plt.xticks(fontsize=13, color='black')
plt.yticks(fontsize=13, color='black')

plt.xlabel('Month', fontsize = 15, color='black')
plt.ylabel('Sea Ice Extent (million square kilometers)', fontsize = 15, color='black')
plt.title('Sea Ice Extent in 1980', fontsize = 15, color='black')
```

Avarage sea ice extent in 1980.

```python
sum = 0
for value in data_1980.values:
    sum += value

average_sea_ice_extent_1980 = sum / 12
```

```python
model_year.append(year)
model_temp_anomaly.append(temp_anom_1980)
model_average_sea_extent.append(average_sea_ice_extent_1980)
```
